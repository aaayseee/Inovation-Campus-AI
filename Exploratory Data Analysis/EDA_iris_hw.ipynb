# 🌸 Exploratory Data Analysis (EDA) with the Iris Dataset

In this notebook, we will perform a step-by-step Exploratory Data Analysis (EDA) using the Iris dataset.

The goal is to learn how to:
- load a dataset
- inspect its structure
- summarize the data
- filter and group rows
- create visualizations
- build simple observations from the results


## 1. Import libraries

In this step, we import the Python libraries that we will use throughout the notebook.

- **pandas** helps us work with tables and datasets.
- **numpy** helps with numerical operations.
- **matplotlib.pyplot** helps us create basic plots.
- **seaborn** helps us create cleaner and more informative visualizations.

We usually import all required libraries at the beginning so they are ready for the rest of the analysis.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load the Iris dataset

Now we load the Iris dataset from the Seaborn library.

This dataset contains flower measurements for three iris species:
- setosa
- versicolor
- virginica

We use `head()` to display the first 5 rows and get an initial idea of what the dataset looks like.


```python
iris = sns.load_dataset("iris")
iris.head()
```

## 3. Check the shape of the dataset

In this step, we check the overall size of the dataset.

The `shape` function gives us two numbers:
- the number of rows
- the number of columns

This helps us quickly understand how large the dataset is.

```python
iris.shape
```

## 4. Show column names

Here we display the names of all columns in the dataset.

This helps us understand what variables are available for analysis.

```python
iris.columns
```

## 5. General information about the dataset

The `info()` function gives a quick overview of the dataset.

It shows:
- column names
- number of non-null values
- data types

This is useful for understanding the structure of the data.

```python
iris.info()
```

## 6. Check missing values

Now we check whether the dataset contains missing values.

This is an important data cleaning step, because missing values can affect later analysis.

```python
iris.isnull().sum()
```

## 7. Display summary statistics for numerical columns

The `describe()` function gives a statistical summary of the numerical columns.

It includes:
- count
- mean
- standard deviation
- minimum
- maximum
- quartiles

This helps us understand the distribution of the numeric features.

```python
iris.describe()
```

## 8. Check unique species values

In this step, we look at the different species names in the dataset.

This helps us see the categories in the target column.

```python
iris['species'].unique()
```

## 9. Count how many samples belong to each species

Now we count how many flowers belong to each species.

This helps us understand whether the dataset is balanced across categories.

```python
iris['species'].value_counts()
```

## 10. Display a few random rows

Instead of always looking at the first rows, we can also display random rows.

This helps us inspect the dataset from different parts.


```python
iris.sample(5)
```

## 11. Select only one column

Here we select only the `sepal_length` column and display its first few values.

This helps students understand how to work with a single column in a DataFrame.


```python
iris['sepal_length'].head()
```

## 12. Basic statistics for one column

Now we calculate some basic statistics for one single column.

We look at:
- mean
- median
- minimum
- maximum

This helps us understand one feature in more detail.

```python
print("Ortalama:", iris['sepal_length'].mean())
print("Medyan:", iris['sepal_length'].median())
print("Min:", iris['sepal_length'].min())
print("Max:", iris['sepal_length'].max())
```

## 13. Filter rows where sepal length is greater than 5

Filtering helps us select only the rows that satisfy a condition.

In this example, we keep only flowers with `sepal_length` greater than 5.

```python
iris[iris['sepal_length'] > 5].head()
```

## 14. Filter only one species

Here we filter the dataset to keep only the flowers of the species `setosa`.

This is useful when we want to focus on one category only.

```python
iris[iris['species'] == 'setosa'].head()
```

## 15. Filter multiple conditions

We can also filter using more than one condition at the same time.

In this example, we select rows where:
- the species is `virginica`
- and the petal length is greater than 5

```python
iris[(iris['species'] == 'virginica') & (iris['petal_length'] > 5)].head()
```

## 16. Group by species and calculate mean

Grouping helps us summarize the data by categories.

Here we group the flowers by species and calculate the mean of each numerical feature.

```python
iris.groupby('species').mean()
```

## 17. Group by species and calculate median

In this step, we again group by species, but this time we calculate the median.

This is useful for comparing the center of each group.


```python
iris.groupby('species').median()
```

## 18. Sort values by petal length

Sorting helps us arrange the dataset in ascending or descending order.

Here we sort the rows by `petal_length` from smallest to largest.

```python
iris.sort_values(by='petal_length')
```

## 19. Histogram of sepal length

A histogram shows how values are distributed.

In this example, we create a histogram of `sepal_length` to see:
- the range of values
- the most common values
- the general shape of the distribution


```python
sns.histplot(iris['sepal_length'], kde=True)
plt.title('Sepal Uzunluğu Dağılımı')
plt.show()
```

## 20. Histogram of petal length

Now we create another histogram, this time for `petal_length`.

This allows us to compare the distribution of petal length values.

```python
sns.histplot(iris['petal_length'], kde=True, color='purple')
plt.title('Petal Uzunluğu Dağılımı')
plt.show()
```

## 21. Count plot for species

A count plot shows how many samples belong to each category.

Here we visualize how many flowers belong to each species.

```python
sns.countplot(x='species', data=iris)
plt.title('Türlere Göre Örnek Sayısı')
plt.show()
```

## 22. Boxplot of sepal length by species

A boxplot helps us compare distributions across categories.

Here we compare `sepal_length` for each species and observe:
- median
- spread
- possible outliers

```python
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title('Türlere Göre Sepal Uzunluğu')
plt.show()
```

## 23. Boxplot of petal length by species

Now we create a boxplot for `petal_length` by species.

This can help us see whether petal length separates species more clearly.


```python
sns.boxplot(x='species', y='petal_length', data=iris)
plt.title('Türlere Göre Petal Uzunluğu')
plt.show()
```

## 24. Scatter plot: sepal length vs sepal width

A scatter plot helps us study the relationship between two numerical variables.

In this example, we compare:
- sepal length
- sepal width

We also color the points by species to see class patterns.


```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Sepal Uzunluğu ve Sepal Genişliği Dağılımı')
plt.show()
```

## 25. Scatter plot: petal length vs petal width

Here we create another scatter plot using:
- petal length
- petal width

This plot is often very useful because the species are usually more clearly separated with petal measurements.


```python
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris)
plt.title('Petal Uzunluğu ve Petal Genişliği Dağılımı')
plt.show()
```

## 26. Pairplot for the whole dataset

A pairplot shows pairwise relationships between all numerical features.

This is a very useful EDA tool because it combines:
- scatter plots between variables
- distributions on the diagonal

We also color the plots by species.

```python
sns.pairplot(iris, hue='species')
plt.show()
```

## 27. Correlation matrix

The correlation matrix shows the strength of the linear relationship between numerical features.

Values close to:
- **1** mean strong positive correlation
- **-1** mean strong negative correlation
- **0** mean weak or no linear correlation


```python
iris.corr(numeric_only=True)
```

## 28. Correlation heatmap

A heatmap gives a visual representation of the correlation matrix.

This makes it easier to quickly identify strong and weak relationships.

```python
sns.heatmap(iris.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Korelasyon Isı Haritası')
plt.show()
```

## 29. KDE plot for numerical features

A KDE plot shows the estimated distribution of numerical variables using smooth curves.

This helps us compare the overall shape of each feature distribution.

```python
sns.kdeplot(data=iris, x='sepal_length', hue='species', fill=True)
plt.title('Sepal Uzunluğu KDE Grafiği')
plt.show()
```

## 30. Average petal length by species as bar chart

Now we calculate the average petal length for each species and visualize it with a bar chart.

This makes group differences easier to interpret.

```python
sns.barplot(x='species', y='petal_length', data=iris)
plt.title('Türlere Göre Ortalama Petal Uzunluğu')
plt.show()
```

## 31. Find the species with the highest average petal length

This step helps us identify which species has the largest average petal length.

We sort the mean values from highest to lowest.

```python
iris.groupby('species')['petal_length'].mean().sort_values(ascending=False)
```

## 32. Show the first 10 rows of virginica flowers

Here we focus on one species again and display the first 10 rows for `virginica`.

This is helpful when students want to inspect one class in more detail.


```python
iris[iris['species'] == 'virginica'].head(10)
```

## 33. Create a new column: petal area

Feature engineering means creating a new variable from existing variables.

In this example, we create a new column called `petal_area` by multiplying:
- petal length
- petal width

This gives a simple approximation of petal size.


```python
iris['petal_area'] = iris['petal_length'] * iris['petal_width']
iris.head()
```

## 34. Compare average petal area by species

Now we compare the average petal area for each species.

This helps us see whether the new feature is useful for distinguishing categories.


```python
iris.groupby('species')['petal_area'].mean()
```

## 35. Visualize petal area by species

Finally, we visualize petal area using a boxplot.

This helps us compare the distribution of the new feature across species.


```python
sns.boxplot(x='species', y='petal_area', data=iris)
plt.title('Türlere Göre Petal Alanı')
plt.show()
```

In the final step, we summarize the main findings from the analysis.

This is important because EDA is not only about writing code. It is also about interpreting the results.
